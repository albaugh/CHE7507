# **LOOPS**

'For loops' are the basic method for traversing data structures or doing repeated computations or procedures.  Below a simple 'for loop' starts from 0 and goes to $N$-1.  In each iteration of the loop the counting variable $i$ gives the loop's current iteration. **Note:** *Indentation is important in Python.  The stuff that is inside the loop gets indented.  The stuff outside the loop is not indented.*

```python
N = 5
for i in range(0,N):
  print('Inside the loop: ',i)

print('Outside the loop!')
```

Note: <code>range(x,y)</code> will count from <code>x</code> to <code>y-1</code>.  If you don't specify  <code>x</code> then it will assume it to be 0.

```python
N = 5
for i in range(N):
  print(i)
```

We can use loops to iterate through elements of a list.

```python
float_list = [-3.1, 4.5, 6.7]
for f in float_list:
  print(f)
```

```python
string_list = ['red','car','run']
for s in string_list:
  print(s)
```

We can use loops to do computations.

```python
sum = 0
for i in range(5):
  sum += i
print(sum)
```

We can nest loops, which means putting one loop inside another.  For example, this is useful for traversing a matrix.

```python
matrix = [[-1.7, 4.9, -2.0], [1.3, -3.0, 5.2], [1.6, 5.3, -2.1]]
for i in matrix: #this will loop over the rows
  for j in i: #this will loop over the elements in each row
    print(j)
```

<code>zip</code> is a useful way to iterate through two lists at the same time.

```python
float_list = [-3.1, 4.5, 6.7]
string_list = ['red','car','run']
for f,s in zip(float_list, string_list):
  print(f,s)
```

<code>zip</code> will stop when it has reached the end of the shortest list.

```python
float_list = [-3.1, 4.5, 6.7, 0.0, 4.1, 5.6]
string_list = ['red','car','run']
for f,s in zip(float_list,string_list):
  print(f,s)
```

<code>enumerate</code> is a useful tool for counting while you iterate through a list.

```python
string_list = ['red','car','run']
for i,s in enumerate(string_list):
  print(i,s)
```

We can even use <code>zip</code> and <code>enumerate</code> together!

```python
float_list = [-3.1, 4.5, 6.7, 0.0, 4.1, 5.6]
string_list = ['red','car','run']
for i,(f,s) in enumerate(zip(float_list,string_list)):
  print(i,f,s)
```

'List comprehensions' are a Python special and they take some getting used to.  If you want to do a bunch of stuff to a list you can do it with a special synthax within the list instead of writing a big <code>for</code> loop.  Say I wanted to count by 3 starting at 0, do this $N$ times, and put the results into a list.  I can use a for loop...

```python
N = 5
list = [] #start with an empty list
for i in range(N):
  list.append(i*3) #append the current data to the list
print(list)
```

...or I could use a Python list comprehension.

```python
N = 5
list = [i*3 for i in range(N)]
print(list)
```

Another type of loop is the 'while loop'.  In a while loop, we continue to iterate the loop while its condition remaines true.  This is useful when you want to continue to do something as long as a certain condition is met.

```python
count = 0
while count < 5:
  count += 1

print(count)
```

Let's look at what's going on inside the loop:

```python
count = 0
while count < 5:
  count += 1
  print('Inside loop: ', count, count < 5)

print(count)
```

# **LOGICAL STATEMENTS**

'If statements' are a way of directing the flow of your program based on certain conditions.  A basic construction is the 'if-else' statement that evaluates a condition based on two possible cases.  **Note:** Python's indenting is important here, too!  After an <code>if</code> or <code>else</code>, anything that is indented will be part of it.  If its not indented, it will not be part of the statement.

```python
number = 3
if number < 5:
  print('Number is less than 5.')
else:
  print('Number is not less than 5.')
```

We can expand our 'if-else' statement to more than two cases by including 'else-if' statements with <code>elif</code>.

```python
number = 12
if number < 3:
  print('Number is less than 3.')
elif number == 6:
  print('Number is 6.')
elif number == 7:
  print('Number is 7.')
elif number < 13:
  print('Number is greater than 7 and less than 13.')
else:
  print('Number is greater than or equal to 13.')
```

# **FUNCTIONS**

Functions are blocks of code that take input and return output.  A principle of good code is that it is D.R.Y. (don't repeat yourself).  If you find yourself writing the same bit of code several times, just make it a function and call that function whenever you need to.  Here's the synthax for a linear function $y = mx + b$, where the input is $x$ and also the slope ($m$) and intercept ($b$), and the output is $y$.  In these notebooks, make sure you place the function definition before any calls you make to it.

```python
def function(x, m, b):
  y = m*x + b
  return y
```

We can now call the function whenever we want.

```python
output = function(1.0, 5.0, -8.0)
print(output)
```

```python
x = -3.0
m = -6.0
b = 10.0
print(function(x, m, b))
```

```python
x = 2.0
m = -0.5
b = 4.2
y = function(x, m, b)
print(y)
```

A function does not need to return a value.  It can also just take inputs and do something with them without returning a value.  Here's a function that takes a list, finds the minimum value, and then prints it.

```python
def min_list(list):
  minimum = list[0]
  for i in list:
    if i < minimum:
      minimum = i
  print(minimum)
```

```python
a = [-3.0, -6.0, -12.0, 15, 26]
min_list(a)
```

A function can also return multiple values and we can collect them in a comma-separated list when we call the function.

```python
def func(x, y, z):
  val1 = x*y + z
  val2 = x + y - z
  message = 'Hello!'
  return val1, val2, message

a, b, c = func(1.0, 2.0, 3.0)
print(a)
print(b)
print(c)
```
